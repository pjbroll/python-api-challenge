### Analysis of weather observations for cities along latitudes from South Pole to North Pole
- Temperatures are warmer near the equator than the poles based on plots of maximum temperature with latitude. The highest temperatures were seen in the Southern Hemisphere as November is a summer month in that hemisphere.
- Humidity values varied randomly from around 10% to 100% - not linked with latitude. Humidity is a function of the amount of moisture in the local airmass and the ambient temperature.  Relative humidity increases as the difference decreases between the ambient temperature and the dewpoint temperature. 
- Cloudiness is also distributed randomly - not linked with latitude. Cloudiness ranged from 0% (clear skies) to 100% (overcast).
- Wind speeds in cities were generally at or below 20 miles per hour. In the far Southern hemispere there is little landmass south of 40 degrees until reaching the continent of Antarctica. This open water region is referred to as the "roaring 40s and 50s". While there are fewer reporting stations this far south the data shows the proportion of higher wind speeds to overall reports is greater than the windspeed distributions at other latitudes. A similar phenomena is observed as you move towards the North pole but the overall effect is less because of increased landmass generated friction or orographic effects.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from time import sleep
# Dependencies
import openweathermapy.core as owm
import urllib.error as ur
from datetime import datetime

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
# print(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Create settings dictionary with information we're interested in
settings = {
    "units": "imperial", 
    "appid": api_key
}
```

```python
# Define lists to hold data
cityName = []
cloudcover =[]
countryCode = [] 
date = []
humidity = [] 
lat = [] 
long = [] 
maxTemp = [] 
windSpeed = []
```

```python
summary = [
    "name", 
    "clouds.all", 
    "sys.country", 
    "dt", 
    "main.humidity", 
    "coord.lat", 
    "coord.lon", 
    "main.temp_max", 
    "wind.speed"
]
```

```python
# Initialize record and set counter
record = 0
set = 1
# Get current weather
print(f"Beginning Data Retrieval")
for city in cities:
    
    try:
        current_weather = owm.get_current(city, **settings)
        # increment record count
        record += 1
        print(f"Processing Record {record} of set {set} | {city}")
        data = current_weather(*summary)
        cityName.append(data[0])
        cloudcover.append(data[1])
        countryCode.append(data[2])
        date.append(data[3])
        humidity.append(data[4])
        lat.append(data[5])
        long.append(data[6])
        maxTemp.append(data[7])
        windSpeed.append(data[8])
        sleep(0.1) # Time in seconds delay before next call
        if record == 50:
            set += 1
            record = 0
    except (NameError, ur.URLError, ur.HTTPError):
        print(f"City not found. Skipping...")

print(f"------------------------")
print(f"Data Retrieval Complete")
print(f"------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create a data frame from cities, and weather values
weather_dict = {
    "City": cityName,
    "Cloudiness": cloudcover,
    "Country": countryCode,    
    "Date": date,
    "Humidity": humidity,
    "Latitude": lat, 
    "Longitude": long, 
    "Max. Temp.": maxTemp,
    "Wind Speed": windSpeed
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
weather_data.to_csv("output/city_weather.csv", index=False, header=True)
```
